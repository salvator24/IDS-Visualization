# Visualizing the NSL-KDD Dataset

The notebook uses TSNE to make visualizations on NSL-KDD Train and Test set in order to recognize outliers in the dataset

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import KernelPCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
% matplotlib inline
```

```python
train_set_2 = pd.read_csv('resources/NSLKDD_10%_2class_preprocessed_normalized/train_10%_nslkdd_2class_preprocessed_normalized.csv')
test_set_2 = pd.read_csv('resources/NSLKDD_10%_2class_preprocessed_normalized/test_10%_nslkdd_2class_preprocessed_normalized.csv')
```

```python
train_set_5 = pd.read_csv('resources/NSLKDD_10%_2class_preprocessed_normalized/train_10%_nslkdd_5class_preprocessed_normalized.csv')
test_set_5 = pd.read_csv('resources/NSLKDD_10%_2class_preprocessed_normalized/test_10%_nslkdd_5class_preprocessed_normalized.csv')
```

```python
train_set_2.shape, test_set_2.shape, train_set_5.shape, test_set_5.shape, 
```

```python
train_y_2 = train_set_2['41']
train_y_5 = train_set_5['41']
train_x = train_set_2.drop(['41'], axis=1)
test_y_2 = test_set_2['41']
test_y_5 = test_set_5['41']
test_x = test_set_2.drop(['41'], axis=1)
```

```python
model = PCA(n_components=2)
model.fit(train_x)
train = model.transform(train_x)
test = model.transform(test_x)
```

### Train set: 2 class with 41 features

```python
plt.scatter(train[:,0], train[:,1], c=train_y_2, alpha=0.5)
```

### Test set: 2 class with 41 features 

```python
plt.scatter(test[:,0], test[:,1], c=test_y_2, alpha=0.5)
```

###  Train set: 5 class with 41 features

```python
plt.scatter(train[:,0], train[:,1], c=train_y_5, alpha=0.5)
```

### Test set: 5 class with 41 features 

```python
plt.scatter(test[:,0], test[:,1], c=test_y_5, alpha=0.5)
```
